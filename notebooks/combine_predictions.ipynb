```python
import  pandas  as  pd
# Load all 4 prediction files and add position column
qb = pd.read_csv("../data/processed/qb_predictions_2026.csv")
qb["position"] = "QB"

rb = pd.read_csv("../data/processed/rb_predictions_2026.csv")
rb["position"] = "RB"

wr = pd.read_csv("../data/processed/wr_predictions_2026.csv")
wr["position"] = "WR"

te = pd.read_csv("../data/processed/te_predictions_2026.csv")
te["position"] = "TE"

# Combine and rename prediction column to match app expectation
combined = pd.concat([qb, rb, wr, te], ignore_index=True)
combined = combined.rename(columns={"predicted_weekly_ppr": "pred_fp_ppr_2026"})

# Sanity check
print(combined.shape)
print(combined["position"].value_counts())
print(combined.head())

# Export
combined.to_csv("../data/processed/all_predictions_2026.csv", index=False)
print("\nExport complete")
```
