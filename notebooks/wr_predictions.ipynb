```python
import pandas as pd
import joblib
import json

# Auto-select best model from metadata
with open("wr_model_metadata.json", "r") as f:
    metadata_wr = json.load(f)

model_name_wr = metadata_wr["model_name"]
print(f"Using {model_name_wr.upper()} model (RF RMSE: {metadata_wr['rf_rmse']:.4f}, XGB RMSE: {metadata_wr['xgb_rmse']:.4f})")

model_wr = joblib.load(f"best_{model_name_wr}_wr_model.joblib")
with open(f"{model_name_wr}_wr_feature_cols.json", "r") as f:
    feature_cols_wr = json.load(f)

# Load prediction data
pred_wr = pd.read_csv("../data/processed/wr_pred_combined.csv")

# Generate predictions
identifiers_wr = pred_wr[["player_id", "player_display_name", "college_flag"]]
X_pred_wr = pred_wr[feature_cols_wr]

predictions_wr = model_wr.predict(X_pred_wr)

results_wr = identifiers_wr.copy()
results_wr["predicted_weekly_ppr"] = predictions_wr.round(2)
results_wr = results_wr.sort_values("predicted_weekly_ppr", ascending=False).reset_index(drop=True)

print(results_wr.head(20))

results_wr.to_csv("../data/processed/wr_predictions_2026.csv", index=False)
print("\nExport complete")
print(f"Total players predicted: {len(results_wr)}")
print(f"NFL players: {(results_wr['college_flag'] == 0).sum()}")
print(f"CFB players: {(results_wr['college_flag'] == 1).sum()}")
```
