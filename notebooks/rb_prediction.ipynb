```python
import pandas as pd
import joblib
import json

# Load model and expected features
model = joblib.load("best_rf_rb_model.joblib")
with open("rf_rb_feature_cols.json", "r") as f:
    feature_cols_rb = json.load(f)

# Load prediction data
pred_rb = pd.read_csv("../data/processed/rb_pred_combined.csv")

print("Prediction set shape:", pred_rb.shape)
print("\nExpected feature cols:")
print(feature_cols_rb)
print("\nPrediction set cols:")
print(pred_rb.columns.tolist())
```

```python
# Separate identifiers from features
identifiers_rb = pred_rb[["player_id", "player_display_name", "college_flag"]]
X_pred_rb = pred_rb[feature_cols_rb]

# Generate predictions
predictions_rb = model.predict(X_pred_rb)

# Attach predictions back to identifiers
results_rb = identifiers_rb.copy()
results_rb["predicted_weekly_ppr"] = predictions_rb.round(2)

# Sort by predicted score descending
results_rb = results_rb.sort_values("predicted_weekly_ppr", ascending=False).reset_index(drop=True)

# Preview top 20
print(results_rb.head(20))

# Export
results_rb.to_csv("../data/processed/rb_predictions_2026.csv", index=False)
print("\nExport complete")
print(f"Total players predicted: {len(results_rb)}")
print(f"NFL players: {(results_rb['college_flag'] == 0).sum()}")
print(f"CFB players: {(results_rb['college_flag'] == 1).sum()}")
```
